```python
#Created a function to first check which plan the user is in and then calculate the profit 
def profit_per_user(row):
    duration = row['duration']
    text_made = row['texts_made']
    mb_used = row['mb_used']
    plan = row['tariff']
    total = 0
    
    if plan == 'surf':
        if duration > 500:
            over_min = (duration - 500) * 0.03
            total += over_min
        elif text_made > 50:
            over_text = (text_made - 50) * 0.03
            total += over_text
        elif mb_used > (15*1024):
            over_data = (mb_used/1024 - 15) * 10
            total += over_data
        return total + 20

    if plan == 'ultimate':
        if duration > 3000:
            over_min = (duration - 3000) * 0.01
            total += over_min
        elif text_made > 1000:
            over_text = (text_made - 1000) * 0.01
            total += over_text
        elif mb_used > (30*1024):
            over_data = (mb_used/1024 - 30) * 7
            total += over_data
        return total + 70


#Created the profit column and applied the function to get the profit amount    
agg_data['profit'] = agg_data.apply(profit_per_user, axis=1)
agg_data.head(10)
```

The average profit from users of Ultimate and Surf calling plans differs.
The average profit from users in NY-NJ area is different from that of the users from other regions.

```python
#Test that avg profit from Surf is the same avg profit as Ultimate.

#Calculate avg profit of Surf
avg_surf_profit = surf_agg['profit'].mean()

alpha = 0.05

results = st.ttest_1samp(ult_agg['profit'], avg_surf_profit)

print('p-value: ', results.pvalue)

if (results.pvalue < alpha ):
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis")
```

```python
#Test that avg profit from ultimate is the same profit as surf.

#Calculate avg profit of Ultimate
avg_ult_profit = ult_agg['profit'].mean()

alpha = 0.05

results = st.ttest_1samp(surf_agg['profit'], avg_ult_profit)

print('p-value: ', results.pvalue)

if (results.pvalue < alpha ):
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis")
```

I chose alpha to be 0.05 since that amount is widely considered to be a good statistically significant amount. 
After testing if the profits differ between Surf and Ultimate, it appears to be true.

```python
#Test profit from NY-NJ area is the same as other areas

#Separate the data by their areas
ny_nj = agg_data.query('city == "New York-Newark-Jersey City, NY-NJ-PA MSA"')
other_area = agg_data.query('city != "New York-Newark-Jersey City, NY-NJ-PA MSA"')

#Calculate avg profit in ny-nj area
ny_nj_avg_profit = ny_nj['profit'].mean()

alpha = 0.05

results = st.ttest_1samp(other_area['profit'], ny_nj_avg_profit)

print('p-value: ', results.pvalue)

if (results.pvalue < alpha ):
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis")
```
